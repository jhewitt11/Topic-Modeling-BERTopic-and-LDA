```python
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
```

```python
data_df = pd.DataFrame(data = np.random.rand(4,2), columns = ['sports', 'education']).transpose()
```

```python
sns.heatmap(data_df).get_figure().savefig('image1.png')
plt.show
```

```python
fig, axes = plt.subplots(4, 1, squeeze = False, gridspec_kw = {'hspace' : 0.5})

for row_num in range(4):
    
    data_df = pd.DataFrame(data = np.random.rand(4,2), columns = ['sports', 'education']).transpose()
    sns.heatmap(data_df, ax = axes[row_num][0])
    
fig.suptitle('That was easy.')
plt.show
```
